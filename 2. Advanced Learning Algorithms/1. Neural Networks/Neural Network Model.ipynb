**Building a neural network**

Capture.PNG

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.activations import sigmoid
from tensorflow.keras import Sequential
```

```python
x = np.array([[200.0,17.0],
              [120.0,5.0],
              [425.0,20.0],
              [212.0,18.0]])

y = np.array([1,0,0,1])
```

```python
model = Sequential([
    Dense(units=3,activation = 'sigmoid'),
    Dense(units=1,activation = 'sigmoid')])
```

```python
model.compile(loss=tf.keras.losses.BinaryCrossentropy())
model.fit(x,y)
```

```python
model.summary()
```

**General implementation of forward propagation**

Now, there were a lot of expressions in the last coding session,
Let's see how you can simplify this to implement forward prop for a more general neural network, rather than hard coding it for every single neuron like we just did.

image.png

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.activations import sigmoid
from tensorflow.keras import Sequential
```

```python
def dense(a_in,W,b):
  units = W.shape[1]
  a_out = np.zeros(units)
  for j in range(units):
    w = W[:,j]
    z = np.dot(w,a_in) + b[j]
    a_out[j] = sigmoid(z)

  return a_out

W = np.array([[1.0,-3.0,5.0],
              [2.0,4.0,-6.0]]) #2*3 matrix

b = np.array([-1.0,1.0,2.0]) #1*2 matrix

a_in = np.array([-2.0,4.0])
```
