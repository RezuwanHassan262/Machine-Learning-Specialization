```python
# load data
dataset = pd.read_csv('diabetes.csv')
dataset
```

```python
X_train,X_test,y_train,y_test = train_test_split(dataset.iloc[:,0:8].values,dataset.iloc[:,8].values,test_size=0.2,random_state = 42)
```

0 - Copy.PNG

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train,y_train)
y_pred = xgb.predict(X_test)
print(y_pred)
print()
accuracy_score(y_test,y_pred)
```
